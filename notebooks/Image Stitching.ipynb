```python
# Before running, make sure you've run pip install opencv-python, pip install imutils

# Import OpenCV etc.
import os
import cv2
import imutils
import numpy as np
import matplotlib.pyplot as plt

imdir = '/Users/hannah/Dropbox/PhD/CMOST/OneDrive_1_6-16-2021'
```

```python
# Create and set up Stitcher instance
stitcher = cv2.Stitcher.create(mode=1) # Scan mode

# Initialize images list
print("Loading images...")
image_paths = os.listdir(imdir)
image_paths.sort()

images = []
for path in image_paths:
    if not path.startswith('.'):
        image = cv2.imread(imdir+'/'+path)
        images.append(image)
    
print("Done: {} images loaded".format(len(images)))
```

```python
# Now stitch images
print("Stitching images...")
status, stitched = stitcher.stitch(images[0:10])

if status == 0:
    print("Done")
else:
    print("Image stitching failed ({})".format(status))
    
# Convert colors from BGR to RGB
stitched = cv2.cvtColor(stitched, cv2.COLOR_BGR2RGB)
```

```python
# Display stitched image
plt.figure()
plt.imshow(stitched)
plt.show()

# Successfully manages to stitch image30 and image31 due to large amount of overlap but doesn't manage the others
```
